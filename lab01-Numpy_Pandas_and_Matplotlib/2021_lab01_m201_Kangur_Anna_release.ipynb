**Task 2. Count 2-grams [1 point]**

For a sequence of tokens `["a_1", "a_2", ..., "a_N"]` **n-gram** is a subsequence of **n** consecutive tokens.

For example: given a sequence of tokens `["a", "b", "c", "d"]` it has:
- Four unigrams: `["a"], ["b"], ["c"], ["d"]`
- Three 2-grams: `["a", "b"], ["b", "c"], ["c", "d"]`
- Two 3-grams: `["a", "b", "c"], ["b", "c", "d"]`
- One 4-gram: `["a", "b", "c", "d"]`

Your task is two find all 2-grams that start with `` token and count their occurences in all addresses.

Firstly, impement a function `get_n_grams`:

````python
def get_n_grams(input_tokens, n):
    return [input_tokens[i:i+n] for i in range(len(input_tokens)-n+1)]
````

````python
assert get_n_grams(["a", "b", "c", "d"], 1) ==  [['a'], ['b'], ['c'], ['d']]
assert get_n_grams(["a", "b", "c", "d"], 2) ==  [['a', 'b'], ['b', 'c'], ['c', 'd']]
assert get_n_grams(["a", "b", "c", "d"], 3) ==  [['a', 'b', 'c'], ['b', 'c', 'd']]
assert get_n_grams(["a", "b", "c", "d"], 4) ==  [['a', 'b', 'c', 'd']]

print("OK, seems fine!")
````

Secondly, define a function that will filter list of n-grams by the first element of an n-gram:

````python
def filter_n_grams(n_grams, lookup_token):
    n_grams = np.array(n_grams)
    return list(list(x) for x in n_grams[n_grams[:,0] == lookup_token])
````

````python
assert filter_n_grams([
    ["a", "b"], 
    ["a", "c"],
    ["b", "d"],
    ["d", "a"]
    ], 
    "a") == [["a", "b"], ["a", "c"]] 

print("Seems fine!")
````

$$A = \frac{1}{x_2-x_1} ,\quad B = \frac{1}{-y_2+y_1} ,\quad C = \frac{x_1}{-x_2+x_1} + \frac{y_1}{y_2-y_1}$$

$$Ax + By + C = 0$$

$$[yp_{i} \leq y \quad AND \quad y < yp_{i-1} ] \quad OR \quad [yp_{i-1} \leq y \quad AND \quad y < yp_{i} ]$$

$$AND \quad  [x \geq \frac{(xp_{i-1} - xp_{i}) \cdot (y - yp_{i-1})}{(yp_{i-1} - yp_{i})} + xp_{i} ]$$

````python
def inPolygon(x, y, xp, yp):
    c=0
    for i in range(len(xp)):
        if (((yp[i]<=y and y<yp[i-1]) or (yp[i-1]<=y and y<yp[i])) and (x > (xp[i-1] - xp[i]) * (y - yp[i]) / (yp[i-1] - yp[i]) + xp[i])):
            c = 1 - c    
    return c
print(inPolygon(100, 0, (-100, 100, 100, -100), (100, 100, -100, -100)))
````

````python
from scipy import stats
rng = np.random.default_rng()
rvs1 = stats.norm.rvs(loc=5, scale=10, size=500, random_state=rng)
rvs2 = stats.norm.rvs(loc=5, scale=10, size=500, random_state=rng)
stats.ttest_ind(rvs1, rvs2)
````

`This is a two-sided test for the null hypothesis that 2 independent samples have identical average (expected) values. This test assumes that the populations have identical variances by default.`

https://pyprog.pro/statistics_functions/var.html

````python
def T_test(a, b):
    da, db = a.var(), b.var()
    print('default1 = ', da)
    print('default2 = ', db)
    if da == db:
        print(stats.ttest_ind(list(a), list(b), equal_var = False))
    else:
        print('Can not do T-test')
````
